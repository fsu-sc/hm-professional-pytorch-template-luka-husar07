```python
# Loading all the model configurations from the config directory
import json
import os

config_dir = '../configs'

config_files = [f for f in os.listdir(config_dir) if f.endswith('.json')]

model_configs = {}

for file in config_files:
    path = os.path.join(config_dir, file)
    with open(path, 'r') as f:
        model_configs[file] = json.load(f)

for name, config in model_configs.items():
    print(name)
    print(config)
```

```python
# Running the training script for all the configurations

import subprocess

config_paths = ['../configs/config.json', '../configs/underfit.json', '../configs/overfit.json', '../configs/optimal.json']    

for path in config_paths:
    subprocess.run(['python3', '../train.py', '-c', path])
```

## Analysis
If you go to the 'Scalars' menu in tensorboard you will be able to see the graphs for the training and validation loss functions. The plots for all of these are to be expected with the different configurations. The light blue line at the top of the graphs is the underfitting config, the darker blue is the original config, the red in the version that overfits and the orange is the optimal configuration for these models. The light blue line (underfit) does not really drop in loss compared to the other versions and this is because it is unable to fit to the data. This configuration has 1 hidden layer with 5 neurons in it, and we are trying to approximate a harmonic function, so naturally, it will underfit to the data. On the other hand the red line (ovefit) drops significantly faster than all the other configurations and then stops training very early, which will likely prove to show poor generalization in the testing process later on. This is all becasue we are using a model with 5 hidden layers, and 100 neurons per layer, which is the maximum amount we could have for both settings. Thirdly we have the darker blue line (default). This line shows that the model is somewhat effective but not extremely. It does get lower than the underfitting version, but it still not great. Finally we have the orange line (optimal). Here we can clearly see the loss drops at a smoother rate than the overfitting config and then plateaus at an appropriately low value. In this case we used 3 hidden layers with 64 neurons per layer. This version will probably prove to show better generalization and accuracy than all the other models.
